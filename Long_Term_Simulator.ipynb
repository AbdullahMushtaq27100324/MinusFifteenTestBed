```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATASIZE = 56
SENSORS = 3
STREAMING_RATE_PER_SECOND = 1
DAYS = 5
def plot_growth(datasize, sensors, streaming_rate_per_second, days):
    seconds_per_day = 24*60*60

    df = pd.DataFrame({'day': range(1,days+1)}) 

    df['size'] = datasize*sensors*streaming_rate_per_second*seconds_per_day*df['day']

    return df

data = plot_growth(DATASIZE, SENSORS, STREAMING_RATE_PER_SECOND, DAYS)
data.plot(x='day',y='size',kind='line',marker='o')
plt.title('Size Growth Rate')
plt.show()
```
